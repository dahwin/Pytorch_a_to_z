```python
import torch.nn as nn
import torch
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = nn.Linear(100, 50)
        self.dropout = nn.Dropout(p=0.5)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = self.fc1(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x
```

```python
import torch
import numpy as np

# Create random integers for dimensions x and y
x = torch.randint(10, (5, 3))
y = torch.randint(10, (3, 5))

# Define the number of embeddings (num_embeddings)
num_embeddings = len(x)

# Initialize an embedding matrix using tensor initialization function from NumPy
embedding = np.random.rand(num_embeddings, y[0], x[0])

# Convert the embedding array to Torch tensor
embedding_tensor = torch.from_numpy(embedding).float().requires_grad_(require_grad=True, gradient_req="add")

# Create the Embedding layer by passing `embedding` as the weight tensor and `embedding_tensor` as the input tensor
encoding = torch.nn.functional.linear_layer(embedding_tensor, embedding)
```

```python
n, d, m = 3, 5, 7
x = torch.randint(10, (5, 3))
y = torch.randint(10, (3, 5))
embedding = nn.Embedding(y,x)
print(embedding)
W = torch.randn((m, d), requires_grad=True)
idx = torch.tensor([1, 2])
a = embedding.weight.clone() @ W.t()  # weight must be cloned for this to be differentiable
b = embedding(idx) @ W.t()  # modifies weight in-place
out = (a.unsqueeze(0) + b.unsqueeze(1))
loss = out.sigmoid().prod()
loss.backward()
```
