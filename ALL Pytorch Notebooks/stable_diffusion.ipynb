```python
%%time
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler
import torch
model_id = "stabilityai/stable-diffusion-2"

# Use the Euler scheduler here instead
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "a super wide shot of 3 skaters, male, diverse, 20s, sitting outside a burger joint eating, and 2 cute girls in their 20s about to sit down next to them, dusk, alexa 35 camera, anamorphic, cinematic, in the style of david fincher, hyper realistic, 4k, ultra-detailed --ar 16:9 --v 4 "
image = pipe(prompt).images[0]
p = ' a super wide shot of 3 skaters, male, diverse, 20s, sitting outside a burger joint eating,'
image.save(f"{p}.png")
```
