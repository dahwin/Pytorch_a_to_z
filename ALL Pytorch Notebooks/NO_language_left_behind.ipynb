```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

text_to_translate = "Life is like a box of chocolates"
model_inputs = tokenizer(text_to_translate, return_tensors="pt").to('cuda')

# translate to French
gen_tokens = model.generate(**model_inputs, forced_bos_token_id=tokenizer.lang_code_to_id["Fr"])
print(tokenizer.batch_decode(gen_tokens, skip_special_tokens=True))
```

```python
from transformers import AutoTokenizer, M2M100ForConditionalGeneration

model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M",torch_dtype="auto",load_in_8bit=True, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained("facebook/m2m100_418M")

text_to_translate = "My blood, sweat, and tears. My body, mind, and soul I know well they're all yours. This is a spell that will punish me"
model_inputs = tokenizer(text_to_translate, return_tensors="pt").to('cuda')

# translate to French
gen_tokens = model.generate(**model_inputs, forced_bos_token_id=tokenizer.get_lang_id("bn"))
print(tokenizer.batch_decode(gen_tokens, skip_special_tokens=True))
```
