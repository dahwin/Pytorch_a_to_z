# Matrix multiplication
**Tow main ways of performing multiplication**



1.   Element-wise multipication
2.   Matrix multiplication

There are two main rules that performing matrix multiplication needs to satisfy:

1. The inner dimensions must match:



*   (3,2) @ (3,2) won't work
*   (2,3) @ (3,2) will work
*   (3,2) @ (2,3) will work

2. The resulting  matrix has the shape of the outer dimensions:

* (2,3) @ (3,2) -> (2,2)
* (3,2) @ (2,3) -> (3,3)




```python
torch.matmul(tensor,tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value+= tensor[i]*tensor[i]
print(value)
```

```python
%%time
torch.matmul(tensor,tensor)
```

# One of the most common erros in deep learning : Shape errors

```python

# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
```

```python
torch.matmul(tensor_A,tensor_B)
```

```python
tensor_B.shape ,tensor_A.shape
```

To fix our tensor shape issues we can manipulate the shape of one of our tensors using a ''transpose''
A transpose switches the axes or dimensions of a given tensor 

# Transpose means coverting columns to rows or rows to columns

```python
tensor_B.T,          tensor_B
```

```python
tensor_B.T
```

```python
tensor_B.T.shape
```

```python
tensor_B.shape
```

**After fixing matric maltiplication**

```python
torch.matmul(tensor_A,tensor_B.T)
```

```python
print(f'Original shapes: tensor_A ={tensor_A.shape}, tensor_B={tensor_B.shape}')
print(f"Make New shapes of a tensor: tensor_A = {tensor_A.shape}, tensor_B.Transpose={tensor_B.T}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

#Reahaping , Stacking, aqueezing and unsqueezing tensor


*   Reshaping - reshapes an input tensor to defined shape
*   View - Resurn a view of an input of certrain shape but keep the same memory as the original tensor


*   Stacking = combine mutiple tensors on top of each other or side by side
*   Squeeze - removes all 1 dimensions from  a tensor


*   Unsqueeze - add a 1 dimension to  a terget tensor
*   Permute - Return a view of the input with dimensions permuted (swapped) in a certain way







```python
tensor_x = torch.arange(1.,10.)
tensor_x, tensor_x.shape
```

```python
# Adding extra dimensions
tensor_x_r = tensor_x.reshape(9,1)
tensor_x_r, tensor_x_r.shape
```

```python
# Change the view
z = tensor_x.view(1,9)
z, z.shape
```

```python
# changing z changes x [becasuse a view of a tensor shares the same memory as the original input]
z[:,0] = 5
z , tensor_x
```

```python
# stack tensore on top of each other
x_stacked = torch.stack([tensor_x,tensor_x,tensor_x,tensor_x],dim=0)
x_stacked
```

```python
tensor_x_r , tensor_x_r.shape
```

```python
# squeexe method basicly remove dimensions
tensor_x_r.squeeze()
```

```python
tensor_x_r.squeeze().shape
```

```python
# torch.squeeze() - remove all single dimensions form a target tensor
print(f"Previous tensor:\n{tensor_x_r}")
print(f"Previous shape:{tensor_x_r.shape}")

# Remove extra dimensions from x reshaped
x_squeezed = tensor_x_r.squeeze()
print(f"\nNew Tensor:{x_squeezed}")
print(f"New shape:{x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim ( dimesion)
print(f"Previous target:{x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor:{x_unsqueezed}")
print(f"New shape:{x_unsqueezed.shape}")
```

```python
# torch.permute - rearrange the dimensions of a target tensor in specifiied order
x_original = torch.rand(size=(224,224,3)) # [height ,wedith,colour_channels]
# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1) # shifs axis 0-> 1, 1->2,2->3
print(f"previous shape:{x_original.shape}")
print(f"New shape:{x_permuted.shape}")
```

# Reproducbility (trying to take random out of random)
In shot how a neural network learns:
start with random numbers-> tensor operations -> update random numbers to try and make them better representaions of the data->again->agin->again
To reduce the randommess in neural networks and Pytorch comes the concept of a random seed
Essentially what the random seed does is 'flavour' the randomness

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
random_seed = 42
torch.manual_seed(random_seed)
random_tensor_C = torch.rand(3,4)
random_tensor_D = torch.rand(3,4)
print(random_tensor_C)
print(random_tensor_C)
print(random_tensor_C ==random_tensor_D)
```
